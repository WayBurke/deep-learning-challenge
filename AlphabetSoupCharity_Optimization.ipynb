## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
import pandas as pd
import tensorflow as tf

# Read the charity_data.csv.

#GitHub link to the CSV file
url = 'https://raw.githubusercontent.com/WayBurke/deep-learning-challenge/main/Resources/charity_data.csv'

application_df = pd.read_csv(url)
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.

application_df= application_df.drop(['EIN','NAME'],axis =1)
application_df
```

```python
# Determine the number of unique values in each column.

application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning

app_type_valueCount = application_df['APPLICATION_TYPE'].value_counts()
app_type_valueCount
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`

application_types_to_replace = list(app_type_valueCount[app_type_valueCount < 500].index)


# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")


# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning

classification_list = application_df['CLASSIFICATION'].value_counts()
classification_list
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1

classification_list[classification_list>1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`

# OPTIMIZATION ATTEMPT: Increasing the list of Classifications

classifications_to_replace = list(classification_list[classification_list <300].index)

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
# One-hot encoding the entire dataframe

application_dummies = pd.get_dummies(application_df)
application_dummies
```

```python
# Split our preprocessed data into our features and target arrays
y = application_dummies.IS_SUCCESSFUL.values
X = application_dummies.drop(columns="IS_SUCCESSFUL").values


# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# OPTIMIZATION ATTEMPT: TO OPTIMIZE THIS MODEL, ADDING AN ADDITIONAL HIDDEN LAYER
# WITH SECOND AND THIRD LAYER HAVING ACTIVATION - RELU
#===========================================================

# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train[0])
hidden_nodes_layer1 =  100
hidden_nodes_layer2 = 40
hidden_nodes_layer3 = 20

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, activation="relu", input_dim=number_input_features))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Third hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer3, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model

nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
import os
#Create for folder callbacks
os.makedirs('optimized_models/', exist_ok=True)

#File Path for checkpoint/callbacks
chk_point_path = "optimized_models/weights-improvement_epoch{epoch:2d}.hdf5"

#Creating/Defining the Callback


save_callback = tf.keras.callbacks.ModelCheckpoint(
    chk_point_path,
    save_weights_only=True, 
    monitor ='val_accuracy',
    mode ='max',
    save_freq='epoch',
    period=50
)
```

```python
# OPTIMIZATION ATTEMPT: Increased the number of epochs to 200
# Train the model

fit_model = nn.fit(X_train_scaled, y_train, epochs=200, callbacks=[save_callback])
```

```python
# Evaluate the model using the test data

model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file

#Create for folder final Model
os.makedirs('h5_models/', exist_ok=True)

nn.save('h5_models/AlphabetSoupCharity_Optimization.h5')
```
